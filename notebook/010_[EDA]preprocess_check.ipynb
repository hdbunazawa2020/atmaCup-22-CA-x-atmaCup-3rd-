```python
from pathlib import Path

def check_paths(df, n=20):
    miss = df["crop_path"].isna().mean()
    print("crop_path missing rate:", miss)
    # å®Ÿãƒ•ã‚¡ã‚¤ãƒ«å­˜åœ¨ãƒã‚§ãƒƒã‚¯ï¼ˆå…ˆé ­nä»¶ã ã‘ï¼‰
    paths = df["crop_path"].dropna().head(n).tolist()
    ok = sum(Path(p).exists() for p in paths)
    print(f"exists (first {n}): {ok}/{len(paths)}")

print("TRAIN")
check_paths(train_pp, n=20)
print("TEST")
check_paths(test_pp, n=20)
```
